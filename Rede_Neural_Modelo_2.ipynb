```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, f1_score, recall_score, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
df = pd.read_csv('Churn_treino.csv', sep=";")
df
```

```python
#Dividindo dados entre treino e teste
X = df.drop("Exited", axis=1)
y = df["Exited"]
```

```python
#Primeiro padronizar os dados e depois o encoder
standardscaler = StandardScaler()
numerical = X.select_dtypes(include=['int64', 'float64']).columns
X[numerical] = standardscaler.fit_transform(X[numerical])
```

```python
X
```

```python
labelencoder = LabelEncoder()
categorical = X.select_dtypes(include=['object']).columns
for col in categorical:
    X[col] = labelencoder.fit_transform(X[col])
```

```python
X
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
modelo = Sequential()
modelo.add(Dense(units=64, activation='relu', input_dim=X_treino.shape[1]))
modelo.add(Dropout(0.4))
modelo.add(Dense(units=32, activation='relu'))
modelo.add(Dropout(0.3))
modelo.add(Dense(units=64, activation='relu'))
modelo.add(Dropout(0.2))
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.fit(X_treino,y_treino,epochs=50, batch_size=32)

```

```python
previsao = modelo.predict(X_teste)
previsao
```

```python
#Transformar os resultado em 0 ou 1
y_pred = (previsao > 0.5).astype('int32')
print(y_pred)
```

```python
print('Acurácia:', accuracy_score(y_teste, y_pred))
print('f1_score:', f1_score(y_teste, y_pred))
print('Recall:', recall_score(y_teste, y_pred))
print('Mariz de Confusao:', confusion_matrix(y_teste, y_pred))
```
