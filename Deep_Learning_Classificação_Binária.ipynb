```python
#atributos previsores
X = pd.read_csv('/content/entradas_breast.csv')
X
```

```python
#classe
y = pd.read_csv('/content/saidas_breast.csv')
y
```

```python
#Divisão da base entre dados de treinamento e dados de teste
from sklearn.model_selection import train_test_split
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size = 0.25)
```

```python
X_treinamento.shape, y_treinamento.shape
```

```python
X_teste.shape, y_teste.shape
```

```python
#Montando a estrutura da rede neural
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
#quantidade de neuronios na camada oculta:
(30+1)/2
```

```python
rede_neural = Sequential([
    #camada de entrada:
    tf.keras.layers.InputLayer(shape=(30,)),
    #camada oculta(nesse exemplo, apenas uma):
    tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
    #Segunda camada oculta:
    tf.keras.layers.Dense(units=16, activation='relu', kernel_initializer='random_uniform'),
    #camada de saída:
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])
#Na camada oculta, o "Dense" significa que a camada oculta está ligada a todos os neurônios da próxima camada
#Units se refere à qtde de neuronios da camada oculta. Para saber a quantidade de neuronios, fazemos a (qtde de atributos previsores + qtd de saídas/2)
#Utilizamos a função relu, pois as outras funções são utilizadas na camada de saída
# A função de ativação da camada de saída será a sigmoide, pois temos um problema de classificação binária
```

```python
rede_neural.summary()
```

```python
#O valor de 496 parametros é o equivalente a 30 neuronios da camada de entrada * 16 neuronios da camada oculta. O restante são os bias adicionados.
# O valor total de parametros correspondem aos pesos que serão ajustados durante a etapa de treinamento
#Em uma rede neural densa, todos os neurônios de uma camada estão ligados a todos os neurônios da próxima camada
```

```python
#Criando um otimizador
#Clipvalue: os pesos nunca vão ultrapassar o valor
otimizador = tf.keras.optimizers.Adam(learning_rate=0.001, clipvalue=0.5)
```

```python
#Aqui, estamos escolhendo o algoritmo que vai ser utilizado para o ajuste dos pesos, também definimos o algoritmo que vai ser usado para calcular os erros (loss)
# LER A DOCUMENTAÇÃO
#Em cada uma das epocas, a rede neural vai processar todos os registros
rede_neural.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
rede_neural.fit(X_treinamento,y_treinamento, batch_size=10, epochs=100)
```

```python
#realizaçõ de previsões da rede Neural e avaliação da base de teste
previsoes = rede_neural.predict(X_teste)
```

```python
previsoes
```

```python
#Converter esses valores de probabilidade em True ou False. Aqui, quanto maior o valor selecionado, maior esperamos a confiança do algoritmo
previsoes = previsoes>0.5
```

```python
previsoes
```

```python
#Agora, precisamos comparar essas previsões com o y_teste (resultado real). O valor de resultado corresponde ao percntual de acerto da rede neural
from sklearn.metrics import confusion_matrix, accuracy_score
accuracy_score(y_teste,previsoes)
```

```python
confusion_matrix(y_teste,previsoes)
#44 tumores benignos foram corretamente classificados
# 9 tumores benignos foram classificados como malignos
# 26 tumores malignos foram classificados como sendo benignos
#64 tumores malignos foram corretamente classificados

#Para calcularmos quantos registros a rede acertou, basta somar a diagonal principal
#Para calcularmos quantos registros a rede errou, basta somar a diagonal secundaria
```
